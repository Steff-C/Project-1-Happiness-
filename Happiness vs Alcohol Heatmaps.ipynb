```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import json
import gmaps
import os

# Import API key
from config import g_key
```

```python
# Load csv and preview
happiness_df = pd.read_csv('output_data/alcohol_vs_happiness.csv')
happiness_df
```

```python
# Target country
country = happiness_df["Country"].to_list()


# Define the endpoint URL
base_url = 'https://maps.googleapis.com/maps/api/geocode/json?'

lat = []
lng = []


for country in country:
 
    
    # Build partial query URL
    query_url = f'{base_url}address={country}&key={g_key}'
    response = requests.get(query_url).json()

    
    
    lat.append(response['results'][0]['geometry']['location']['lat'])
    lng.append(response['results'][0]['geometry']['location']['lng'])
    



```

```python

# Fix Lat/Lng for Georgia - the country

lat[67] = 42.3154
lng[67] = 43.3569

consumption = happiness_df["All types"].to_list()
country = happiness_df["Country"].to_list()
happiness = happiness_df["Score"].to_list()
happy_dict = {
    "Country": country,
    "Lat": lat,
    "Lng": lng,
    "Alcohol Consumption": consumption,
    "Happiness": happiness
   
}
happy_data = pd.DataFrame(happy_dict)

top = happy_data.sort_values(by="Country", ascending=False)

top.tail(50)
```

```python
# Configure gmaps 
gmaps.configure(api_key=g_key)
```
